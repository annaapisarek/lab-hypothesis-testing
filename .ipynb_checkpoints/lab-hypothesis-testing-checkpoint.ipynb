**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df.head(50)
#df["Type 2"].value_counts()
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
df.isnull().sum()
```

```python
#Null Hypothesis (H0): The average HP of Dragon Pokémon is less than or equal to that of Grass Pokemon.
#Alternative Hypothesis (H1): The average HP of Dragon Pokémon is greater than that of Grass Pokemon.

df_Dragon = df[df["Type 1"]=="Dragon"]["HP"]
df_Grass = df[df["Type 1"]=="Grass"]["HP"]

st.ttest_ind(df_Dragon,df_Grass, alternative='greater')
```

```python
#The null hypothesis can be rejected due to low p-value, however this provides strong evidence that Dragon-type Pokémon, on average, have significantly higher HP than Grass-type Pokémon.
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
df_Legendary = df[df["Legendary"]== 1]
df_Non_Legendary = df[df["Legendary"]== 0]

t_stat_hp, p_value_hp = st.ttest_ind(df_Legendary['HP'], df_Non_Legendary['HP'])
t_stat_attack, p_value_attack = st.ttest_ind(df_Legendary['Attack'], df_Non_Legendary['Attack'], equal_var=False)
t_stat_defense, p_value_defense = st.ttest_ind(df_Legendary['Defense'], df_Non_Legendary['Defense'], equal_var=False)
t_stat_sp_atk, p_value_sp_atk = st.ttest_ind(df_Legendary['Sp. Atk'], df_Non_Legendary['Sp. Atk'], equal_var=False)
t_stat_sp_def, p_value_sp_def = st.ttest_ind(df_Legendary['Sp. Def'], df_Non_Legendary['Sp. Def'], equal_var=False)
t_stat_speed, p_value_speed = st.ttest_ind(df_Legendary['Speed'], df_Non_Legendary['Speed'], equal_var=False)

import statsmodels.stats.multitest as smm

# List of p-values from the t-tests for each stat
p_values = [p_value_hp, p_value_attack, p_value_defense, p_value_sp_atk, p_value_sp_def, p_value_speed]

# Apply Bonferroni correction
_, pvals_bonferroni, _, _ = smm.multipletests(p_values, alpha=0.05, method='bonferroni')

# Apply Holm-Bonferroni correction
_, pvals_holm, _, _ = smm.multipletests(p_values, alpha=0.05, method='holm')

print("Bonferroni adjusted p-values:", pvals_bonferroni)
print("Holm adjusted p-values:", pvals_holm)
```

```python
#Based on the adjusted p-values, we can confidently reject the null hypothesis for all the tested stats. This means that Legendary Pokemon do indeed have statistically different values for HP, Attack, Defense, Sp. Atk, Sp. Def, and Speed compared to Non-Legendary Pokémon.
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import math
import pandas as pd

def euclidean_school(row):
    return math.sqrt((-118 - row['longitude'])**2 + (37 - row['latitude'])**2)

def euclidean_hospital(row):
    return math.sqrt((-122 - row['longitude'])**2 + (34 - row['latitude'])**2)

# Apply the functions to create new columns
df['distance_to_school'] = df.apply(euclidean_school, axis=1)
df['distance_to_hospital'] = df.apply(euclidean_hospital, axis=1)

# Display the updated DataFrame
print(df)

    
```

```python
mean_school = df['distance_to_school'].mean()
mean_hospital = df['distance_to_hospital'].mean()

mean_school, mean_hospital
```

```python
close_to_school = df[df["distance_to_school"] <= 3.48]['median_house_value']
far_to_school = df[df["distance_to_school"] < 3.48]['median_house_value']

close_to_hospital = df[df["distance_to_school"] <= 4.06]['median_house_value']
far_to_hospital = df[df["distance_to_school"] > 4.06]['median_house_value']

```

```python
st.ttest_ind(close_to_school, far_to_school, alternative='greater')
```

```python
st.ttest_ind(close_to_hospital, far_to_hospital, alternative='greater')
```

```python
#Means are equal far and close to school or hospital, there is not statistical evidence of a price difference.
```
